```python
image = cv2.imread('/home/kunal/Documents/Hypertensive_Retinopathy/brvo (2).jpg')
color = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
#shifted = cv2.pyrMeanShiftFiltering(gray, 21, 51)
plt.imshow(color)
```
